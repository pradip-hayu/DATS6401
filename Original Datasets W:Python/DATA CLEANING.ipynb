# DATA EXTRACTION AND CLEANING

```python
# Loading required libraries
import pandas as pd
import numpy as np
```

# AIRLINES DELAY DATASET

```python
airlines_delay = pd.read_csv('airlines.csv', header = 0)
airlines_delay
```

```python
#Selecting the following 13 columns from the above dataset:
# (1)# of Delays.Carrier (2)# of Delays.Late Aircraft (3)# of Delays.National Aviation System,
# (4)# of Delays.Security (5)# of Delays.Weather (6)Cancelled (7)Delayed (8)Diverted (9)On Time (10)Year (11)Month
#(12) Month Name (13)Flights.Total
airlines_delay1 = airlines_delay[['# of Delays.Carrier', '# of Delays.Late Aircraft', 
                                  '# of Delays.National Aviation System', '# of Delays.Security',
                                  '# of Delays.Weather', 'Cancelled', 'Delayed', 'Diverted', 'On Time', 'Year', 
                                  'Month', 'Month Name', 'Flights.Total'  ]]
airlines_delay1.head()
```

```python
#Selecting the datas for the year 2004 to 2015 from the above dataframe
airlines_delay2 = airlines_delay1.loc[(airlines_delay1['Year'] != 2003) & (airlines_delay1['Year'] != 2016)]
airlines_delay2.head() #'2004_to_2015' tab of the 'air_delay_dfs.xlsx' file
```

```python
#Grouping the airlines delay data for 2004 to 2015 by Year and Month Name
airlines_delay3 = airlines_delay2.groupby(['Year','Month Name'])['# of Delays.Carrier',
                                                                 '# of Delays.Late Aircraft', 
                                                                 '# of Delays.National Aviation System',
                                                                 '# of Delays.Security', 
                                                                 '# of Delays.Weather','Delayed','Diverted',
                                                                 'Cancelled','On Time','Flights.Total'].sum()
airlines_delay3 # 'year_and_month' tab of the 'air_delay_dfs.xlsx' file
```

```python
#Grouping the airlines delay data for 2004 to 2015 by Year
airlines_delay4 = airlines_delay2.groupby(['Year'])['# of Delays.Carrier','# of Delays.Late Aircraft',
                                                    '# of Delays.National Aviation System', 
                                                    '# of Delays.Security', '# of Delays.Weather','Delayed',
                                                    'Diverted','Cancelled','On Time','Flights.Total'].sum()
airlines_delay4 #'delay_by_year' tab of the 'air_delay_dfs.xlsx' file
```

```python
# Transposing the airlines delay by year
airlines_delay4T = airlines_delay4.T
airlines_delay4T
```

```python
#Grouping the airlines delay data for 2004 to 2015 by Month Name
airlines_delay5 = airlines_delay2.groupby('Month Name')['# of Delays.Carrier','# of Delays.Late Aircraft', 
                                                        '# of Delays.National Aviation System',
                                                        '# of Delays.Security', '# of Delays.Weather','Delayed'
                                                        ,'Diverted','Cancelled','On Time','Flights.Total'].sum()
airlines_delay5 #'delay_by_month' tab of the 'air_delay_dfs.xlsx' file
```

```python
#Creating the cleaned Excel file for the airlines delay data which consists of data for the year 2004 to 2015 for
#the selected columns, airlines delay data for the year 2004 to 2015 grouped by year and month, grouped by year, 
#and grouped by month respectively
writer = pd.ExcelWriter('AIRLINES_DELAY.xlsx')
airlines_delay2.to_excel(writer,'2004_to_2015')
airlines_delay3.to_excel(writer,'year_and_month')
airlines_delay4.to_excel(writer,'delay_by_year')
airlines_delay5.to_excel(writer,'delay_by_month')
writer.save()
```

# AIRLINES SAFETY DATASET

```python
airlines_safety = pd.read_csv('airline-safety.csv', header = 0, index_col = 0)
airlines_safety.head() 
```

```python
#Just getting the incidents, fatal accidents, and fatalities between 2000 to 2014 by dropping 1985 to 1999 data
airlines_safety_00_14 = airlines_safety.drop(['incidents_85_99', 'fatal_accidents_85_99', 'fatalities_85_99'],
                                            axis = 1)
airlines_safety_00_14.head()
```

```python
#Creating the cleaned Excel file for the airlines safety data which consists of the entire data in one tab, and
#just the data of the years 2000 through 2014 in another tab
writer = pd.ExcelWriter('AIRLINES_SAFETY_1.xlsx')
airlines_safety.to_excel(writer,'1985-2014')
airlines_safety_00_14.to_excel(writer,'2000-2014')
writer.save()
```

```python
##DIVIDNG AIRLINES SAFETY INTO VARIOUS LEVELS USING DIFFERENT CRITERIAS (2000 - 2014 data)
#Airlines Safety_Level_A includes airlines with no incidents, no fatal accidents, and no fatalities
airlines_safety_00_14_A= airlines_safety_00_14.loc[(airlines_safety_00_14['incidents_00_14']== 0) & 
                                           (airlines_safety_00_14['fatal_accidents_00_14']== 0) & 
                                           (airlines_safety_00_14['fatalities_00_14']== 0)]
airlines_safety_00_14_A
                                                                                                      
```

```python
#Airlines Safety_Level_B includes airlines with no fatal accidents, and no fatalities, but have incidents
airlines_safety_00_14_B= airlines_safety_00_14.loc[(airlines_safety_00_14['fatal_accidents_00_14']== 0) &
                                           (airlines_safety_00_14['fatalities_00_14']== 0)]
airlines_safety_00_14_B       
```

```python
#Airlines Safety_Level_C includes airlines with fatalities less than 10
airlines_safety_00_14_C= airlines_safety_00_14.loc[(airlines_safety_00_14['fatalities_00_14'] < 10) & 
                                       (airlines_safety_00_14['fatalities_00_14'] != 0)]
airlines_safety_00_14_C
```

```python
#Airlines Safety_Level_D includes airlines with fatalities less than 50
airlines_safety_00_14_D = airlines_safety_00_14.loc[(airlines_safety_00_14['fatalities_00_14'] < 50) & 
                                             (airlines_safety_00_14['fatalities_00_14'] != 0)]
airlines_safety_00_14_D
```

```python
#Airlines Safety_Level_E includes airlines with fatalities less than 100
airlines_safety_00_14_E= airlines_safety_00_14.loc[(airlines_safety_00_14['fatalities_00_14'] < 100) & 
                                               (airlines_safety_00_14['fatalities_00_14'] != 0)]
airlines_safety_00_14_E
```

```python
#Airlines Safety_Level_F includes airlines with fatalities greater than 100
airlines_safety_00_14_F = airlines_safety_00_14.loc[(airlines_safety_00_14['fatalities_00_14'] > 100) &
                                                (airlines_safety_00_14['fatalities_00_14'] != 0)]
airlines_safety_00_14_F
```

```python
#Creating the cleaned Excel file for the airlines safety data which consists of safety levels A to F as mentioned 
#above
writer = pd.ExcelWriter('AIRLINES_SAFETY_2.xlsx')
airlines_safety_00_14_A.to_excel(writer,'Safety_Level_A')
airlines_safety_00_14_B.to_excel(writer,'Safety_Level_B')
airlines_safety_00_14_C.to_excel(writer,'Safety_Level_C')
airlines_safety_00_14_D.to_excel(writer,'Safety_Level_D')
airlines_safety_00_14_E.to_excel(writer,'Safety_Level_E')
airlines_safety_00_14_F.to_excel(writer,'Safety_Level_F')
writer.save()
```

# AIRLINES REVENUE DATASET

# AIRLINES PASSENGER DATASET

```python
airlines_passenger = pd.read_csv('Air_Traffic_Passenger_Statistics.csv')
airlines_passenger
```

```python
#Getting the sum of passenger count based on GEO Summary (Domestic, International) and Geo Region(US, Europe, Asia,
#Canada, Mexico, Middle East)
airlines_passenger['Passenger Count'][airlines_passenger['GEO Summary'] == 'Domestic'].sum()
airlines_passenger['Passenger Count'][airlines_passenger['GEO Summary'] == 'International'].sum()
airlines_passenger['Passenger Count'][airlines_passenger['GEO Region'] == 'US'].sum()
airlines_passenger['Passenger Count'][airlines_passenger['GEO Region'] == 'Europe'].sum()
airlines_passenger['Passenger Count'][airlines_passenger['GEO Region'] == 'Asia'].sum()
airlines_passenger['Passenger Count'][airlines_passenger['GEO Region'] == 'Canada'].sum()
airlines_passenger['Passenger Count'][airlines_passenger['GEO Region'] == 'Mexico'].sum()
airlines_passenger['Passenger Count'][airlines_passenger['GEO Region'] == 'Middle East'].sum()
```

```python
#Sorting Passenger count in descending order
airlines_passenger.sort_values( by = 'Passenger Count', ascending = False)
```

```python
#Grouping Published Airline by Passenger Count
airlines_passenger1 = airlines_passenger.groupby('Published Airline')['Passenger Count'].sum()
airlines_passenger1
```

```python
#Sorting Published Airline based on descending order of Passenger Count
airlines_passenger2=airlines_passenger.groupby('Published Airline')['Passenger Count'].sum().sort_values(ascending
                                                                                                           = False)
airlines_passenger2
```

```python
#Grouping Geo Summary & Published Airline by Passenger Count
airlines_passenger3 = airlines_passenger.groupby(['GEO Summary','Published Airline'])['Passenger Count'].sum()
airlines_passenger3
```

```python
#Grouping Geo Summary, Geo Region & Published Airline by Passenger Count
airlines_passenger4 = airlines_passenger.groupby(['GEO Summary','GEO Region',
                                                  'Published Airline'])['Passenger Count'].sum()
airlines_passenger4
```

```python
#Sorting Passenger Count in descending order
airlines_passenger.groupby(['GEO Summary','GEO Region',
                            'Published Airline'])['Passenger Count'].sum().sort_values(ascending = False)
```

```python
##Creating the cleaned Excel file for the airlines passengers data 
writer = pd.ExcelWriter('AIRLINES_PASSENGER.xlsx')
airlines_passenger1.to_excel(writer,'Airlines Passenger')
airlines_passenger2.to_excel(writer,'Airlines Passenger Descending')
airlines_passenger3.to_excel(writer,'Passengers Geo')
airlines_passenger4.to_excel(writer,'Passengers Geo Region')
writer.save()
```
